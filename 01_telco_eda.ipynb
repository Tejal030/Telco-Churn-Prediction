## STEP 3 — DATA CLEANING

## STEP 3.1—Dropping customerID

```python
df.drop("customerID", axis=1, inplace=True)
```

### 3.2 Fix TotalCharges column (convert to numeric)


```python
# Convert to numeric, force errors into NaN
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

# Fill missing values (created by conversion) with the median
df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].median())
```

### 3.3 Simplify categories: Replace "No internet service" with "No"

```python
cols = ["OnlineSecurity", "OnlineBackup", "DeviceProtection", 
        "TechSupport", "StreamingTV", "StreamingMovies"]

for col in cols:
    df[col] = df[col].replace("No internet service", "No")
```

### 3.4 Simplify categories: Replace "No phone service" with "No"

```python
df["MultipleLines"] = df["MultipleLines"].replace("No phone service", "No")
```

### 3.5 Convert the target column 'Churn' into numeric (Yes=1, No=0)

```python
df["Churn"] = df["Churn"].replace({"Yes": 1, "No": 0})
```

### 3.6 Final check for missing values

Why This Step Is Important for us: 

Machine learning models cannot handle missing values unless you use special algorithms.

Visualizations may give wrong insights if missing data hides patterns.

Before moving to EDA (graphs) and modeling, you must ensure the dataset is complete and consistent.

This step confirms:

All cleaning steps worked correctly

Every column is ready for analysis

No NaN values were left behind during conversions

A dataset with 0 missing values is considered clean and ready for the next stage: Exploratory Data Analysis (EDA).

```python
df.isna().sum()
```

## 4. Univariate Analysis
### 4.1 Categorical Variables
#### Goal:
To Understand how often each category appears (Yes/No, types, etc.) and spot any imbalances or patterns.


```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_cols = df.select_dtypes(include=["object", "category"]).columns

for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    ax = sns.countplot(data=df, x=col)
    
    # Add counts on top of bars
    for p in ax.patches:
        ax.annotate(str(p.get_height()),
                    (p.get_x() + p.get_width() / 2, p.get_height()),
                    ha='center', va='bottom')
    
    plt.title(f"Distribution of {col}", fontsize=14)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

## 4.1 Univariate Analysis—Categorical Features

### Why This Approach?
Most features in the Telco dataset are categorical, meaning they take values like:
- Yes / No  
- DSL / Fiber Optic / None  
- Month-to-month / One-year / Two-year  

To understand how customers behave, it’s important to see:
- Which categories are most common?  
- Whether categories are balanced or imbalanced  
- Whether certain categories dominate  
- Whether data is clean or messy (extra categories, typos, etc.)

Using countplots is the best approach for categorical data because:
- They show the exact frequency of each category.  
- They help identify an imbalance.  
- They reveal patterns at a glance.  
- They help you understand the distribution of customer behaviors.  

This analysis helps guide feature engineering and later modeling.



### What Did I Find from the Categorical Plots?

After visualizing each categorical column, I observed the following key insights:

#### Gender is balanced
- A similar number of male and female customers  
- This feature will likely have low predictive power for churn.  

#### SeniorCitizen is imbalanced
- The majority are not senior citizens
- Senior citizens may churn differently → check later in bivariate analysis  

#### Partner & Dependents show strong imbalance
- Most customers do not have dependents.  
- This may influence churn—customers with families may be more stable.  

#### Contract types are highly imbalanced.
- A large majority use Month-to-month contracts  
- Very few customers have 1-year or 2-year contracts.  
- This is extremely important because:
  - Month-to-month customers can leave easily → high churn risk  
  - Long-term contracts lock customers in → low churn risk  

This is one of the strongest predictors in churn studies.

#### PaymentMethod shows meaningful variation
- Many customers use Electronic Check.  
- Electronic check users often show high churn in churn studies.  
- Credit card & automatic bank transfer users tend to churn less.  

#### Internet service is uneven
- Many customers use fiber optic → may have higher dissatisfaction  
- Customers with no internet service likely have different churn behavior.  

#### Security, Tech Support, Streaming TV, etc.
- Most are No
- This suggests customers are opting out of multiple add-on services.  
- Lack of security or tech support often correlates with higher churn.  



### Why These Findings Matter

Understanding category frequency helps in:

#### Feature Engineering
- We know which categories need encoding.  
- We avoid the dummy variable trap.  
- We drop categories with very low counts.  

#### Predictive Modeling
- Highly imbalanced categories can bias the model.  
- Dominant categories need correct encoding.  
- Some categorical features (like Contract) have a huge impact on churn.  

#### Business Insights
- Identify customer segments more likely to churn  
- Helps build retention strategies  
- Improves the explainability of the churn model  



###  Summary
By using countplots, I was able to gain a deeper understanding of the dataset's composition.  
This step builds the foundation for:

- Bivariate analysis  
- Feature engineering  
- Model selection  
- Interpretation of churn drivers  

Categorical univariate analysis gives a clear picture of how customers behave and what factors might influence their decision to stay or leave.


## 4.2 Univariate Analysis—Numerical Features

### Goal:
To understand how numerical variables are distributed, whether they contain outliers, and what transformations or scaling may be needed before modeling.


```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

for col in numeric_cols:
    plt.figure(figsize=(14, 4))
    
    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True)
    plt.title(f"Histogram of {col}")
    
    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    
    plt.tight_layout()
    plt.show()
```

## Why I chose this step before modeling: I needed to know

1. Whether normalization/scaling is needed

2. Whether outliers could affect the model

3. Whether transformations (log, sqrt, etc.) will help

4. How customer behavior looks numerically

## Findings From Numerical Univariate Analysis

### Tenure (Months the customer has stayed)
Distribution Insight:
- The histogram shows a right-skewed distribution with many customers having low tenure (0–20 months).
- Only a smaller group of customers stay for a long period (50+ months).

What this means:
- The company has a large number of new customers.
- New customers are much more likely to churn because they are not yet committed.
- Long-term customers tend to be stable and loyal.

Why this matters for modeling:
- Tenure will likely be one of the strongest predictors of churn.
- The skew suggests normalization might help certain ML algorithms.


### MonthlyCharges
Distribution Insight:
- Monthly charges are spread fairly evenly, but with slight peaks around:
  - $20–30 (low-cost customers)
  - $70–100 (high-cost customers, often fiber-optic users)

What this means:
- There are two major customer groups:
  - Low-paying basic service users**
  - High-paying premium/fiber users

Churn implications:
- Higher charges may lead to higher churn if customers feel the service is too expensive.
- Lower charges often represent basic plan users who may churn less.

Why this matters for modeling:
- MonthlyCharges is likely an important variable.
- It may also correlate with the type of internet service.




### TotalCharges (Total money paid to the company)
Distribution Insight:
- The distribution is right-skewed, with many customers having low total charges.
- Customers with very high TotalCharges have stayed for years.

Key interpretation:
- TotalCharges is essentially a “lifetime value” metric.
- Customers with low TotalCharges are often:
  - New customers
  - High churn risk

Why this matters:
- TotalCharges acts like tenure but includes the cost aspect.
- Strong candidate for important predictive power.

Note:
We cleaned this column earlier because it had incorrect types. Now the distribution clearly reflects customer tenure + spend.


### Outliers
Boxplot insights:
- MonthlyCharges has some higher-value outliers (premium services), but they are legitimate data points.
- TotalCharges has high-spend outliers—these represent long-term loyal customers.
- No unrealistic or dirty outliers were found after cleaning.

What this means:
- No removal of outliers is needed.
- These represent genuine customer segments.

## 5. Bivariate Analysis
### 5.1 Categorical Variables vs Churn
#### Goal:
To examine how different categories (Yes/No, types, options) affect customer churn. This helps identify which customer groups are more likely to leave.


```python
import seaborn as sns
import matplotlib.pyplot as plt

categorical_cols = df.select_dtypes(include=["object", "category"]).columns

for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col, hue="Churn")
    plt.title(f"{col} vs Churn", fontsize=14)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

### Expected Insights

#### Contract Type
- Month-to-Month customers show very high churn.
- One-year and Two-year** contract customers churn much less.
- This is one of the strongest predictors of churn.

#### OnlineSecurity, TechSupport
- Customers without online security or tech support churn more.
- Customers with these services churn much less.

#### Tenure Groups
- Customers with low tenure (new customers) churn more.
- Long-term customers rarely churn.

#### PaymentMethod
- Electronic check users have the highest churn.
- Automatic payments show lower churn.

#### InternetService
- Fiber optic customers churn more—likely because of higher monthly cost.

#### MultipleLines, Dependents, Partner
- Dependents/Partner customers churn less → more stable group.
